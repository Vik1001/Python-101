```python
restaurant_menu={'sandwich':100,'shawarma':120,'tea':10,'butter biscuit': 5,'milkshake':70}
total=0
while True:  
    order=input('enter order: ')
    if not order:
        print(f'bill amount is {total}')
        break
    if order in restaurant_menu:
        total+=restaurant_menu[order]
        print(f'{order} costs {restaurant_menu[order]}, total is {total}')
    else:
        print('we do not carry this product please try again')
```

```python
#how to check if a given input is a key in a dictionary
d={'a':1,'b':2,'c':3}
key='z'
if key in d:
    print(f'{key}\'s value is {d[key]}')
else:
    print(f'{key} doesn\'t exist in d')
```

```python
# d.keys(),d.values(),d.items
d={'a':1,'b':2,'c':3,'x':5,'y':200}
print(f"d\'s keys are {d.keys()}")
print(f"d\'s values are {d.values()}")
print(f'd\'s items are {d.items()}')
```

```python
#we can iterate through a dictionary using dictionary.items 

for key,value in d.items():
    print(f'{key}:{value}')
```

```python
d={}
while True:
    city=input('City: ').strip().capitalize()
    if not city:
        break
    rain=int(input('Rain: ').strip())
    if city not in d:
        d.setdefault(city,rain)
    else:
        d[city]=d[city]+rain
for key,value in d.items():
    print(f'{key} : {value}')
```
