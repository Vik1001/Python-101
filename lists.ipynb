```python
mylist=[10,20,30]
mylist = mylist.append(40) # def append() -- returns None changes are done to the list
print(mylist)
type(mylist)
```

In the given code, mylist is assigned the value returned by mylist.append(40). However, the append() method of a list in Python does not return a new list. Instead, it modifies the existing list by adding the element passed as an argument to it and returns None.

Therefore, when mylist.append(40) is executed, mylist is modified to [10, 20, 30, 40] and the return value of mylist.append(40) is None. This None value is then assigned back to mylist, replacing the original list with None.

methods that change data return None.

```python
database= [['abc','def','ghi'],
           ['jkl','mno','pqr'],
           ['stu','vwx','yz']]
output=''
for one_record in database:
    output+='\t'.join(one_record) + '\n'
print(output)
```

```python
# the above method is a memory intensive method as strings are immutable. Output is assigned a new string every time the loop 
# runs. Rather take output as a list and since lists are mutable only new elements are appended to the list. It isn't as memory 
# intensive as the above method

database= [['abc','def','ghi'],
           ['jkl','mno','pqr'],
           ['stu','vwx','yz']]
output=[]
for one_record in database:
    output.append('\t'.join(one_record))
print('\n'.join(output))
```
